```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.simplefilter(action='ignore')
%config InlineBackend.figure_format = 'retina'
plt.rcParams['figure.figsize'] = (6, 4)
```

```python
# CarSeat Îç∞Ïù¥ÌÑ∞
carseat = pd.read_csv('https://raw.githubusercontent.com/Jangrae/csv/master/Carseats.csv')
carseat.drop('Education', axis=1, inplace=True)
carseat.head()
```

- Population: ÏßÄÏó≠ Ïù∏Íµ¨Ïàò(Îã®ÏúÑ: 1,000Î™Ö)

## **2. Î∂ÑÏÑù**

- carseat Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏóêÏÑú ÏïÑÎûò Î≤îÏ£ºÌòï Î≥ÄÏàòÎ•º Îã§Ïùå Î∞©Î≤ïÏúºÎ°ú Î∂ÑÏÑùÌïòÏÑ∏Ïöî.
    - ÏàòÏπòÌôî: Î≤îÏ§èÍ∞í Í∞úÏàò Î∞è ÎπÑÏú®
    - ÏãúÍ∞ÅÌôî: sns.countplot() Ìï®Ïàò, plt.pie() Ìï®Ïàò

**1Ô∏è‚É£ ShelveLoc**

```python
# ÎπàÎèÑÏôÄ ÎπÑÏú®
print("ShelveLoc ÎπàÎèÑ:")
print(carseat['ShelveLoc'].value_counts())
print("\nShelveLoc ÎπÑÏú®:")
print(carseat['ShelveLoc'].value_counts(normalize=True), "\n")
```

```python
# ÏãúÍ∞ÅÌôî
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.countplot(x='ShelveLoc', data=carseat, palette='Set2')
plt.title("Countplot")

plt.subplot(1,2,2)
carseat['ShelveLoc'].value_counts().plot.pie(autopct="%.1f%%", startangle=90, colors=sns.color_palette("Set2"))
plt.ylabel('')
plt.title("Pie Chart")

plt.show()
```

**2Ô∏è‚É£ Urban**

```python
# ÎπàÎèÑÏôÄ ÎπÑÏú®
print("Urban ÎπàÎèÑ:")
print(carseat['Urban'].value_counts())
print("\nUrban ÎπÑÏú®:")
print(carseat['Urban'].value_counts(normalize=True), "\n")
```

```python
# ÏãúÍ∞ÅÌôî
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.countplot(x='Urban', data=carseat, palette='Set2')
plt.title("Countplot")

plt.subplot(1,2,2)
carseat['Urban'].value_counts().plot.pie(autopct="%.1f%%", startangle=90, colors=sns.color_palette("Set2"))
plt.ylabel('')
plt.title("Pie Chart")

plt.show()
```
